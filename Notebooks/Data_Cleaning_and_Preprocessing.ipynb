```python
# Function to fix invalid times (24:xx:xx to 00:xx:xx) - Will not be used for 'Date of Birth'
def fix_invalid_times(date_str):
    if isinstance(date_str, str):  # Only apply to strings
        # Check for '24' hour and replace it with '00' of the next day
        date_str = re.sub(r'24:(\d{2}:\d{2}:\d{2})', r'00:\1', date_str)
    return date_str

# Normalize the date format: MM/DD/YYYY -> YYYY-MM-DD
def normalize_date_format(date_str):
    try:
        # If format is like 'MM/DD/YYYY', change it to 'YYYY-MM-DD'
        if re.match(r'\d{2}/\d{2}/\d{4}', date_str):  # Matches MM/DD/YYYY
            return pd.to_datetime(date_str, format='%m/%d/%Y').strftime('%Y-%m-%d')
        # If already in 'YYYY-MM-DD', return as is (drop time part)
        return pd.to_datetime(date_str).strftime('%Y-%m-%d')
    except Exception as e:
        # Return the original string if any error occurs
        return date_str


# Separate function to handle Date of Birth column
def process_date_of_birth(df_copy, column_name='date_of_birth'):
    """
    Function to process the 'Date of Birth' column to normalize the date format: MM/DD/YYYY -> YYYY-MM-DD
    """
    # Normalize the date format: MM/DD/YYYY -> YYYY-MM-DD
    df_copy[column_name] = df_copy[column_name].apply(normalize_date_format)

    # Convert the column to datetime, catching invalid entries
    df_copy[column_name] = pd.to_datetime(df_copy[column_name], errors='coerce')

    # Return the updated dataframe
    return df_copy

# Example usage:
# Assuming df_copy is your DataFrame and 'Date of Birth' is the column you want to process
df_copy = process_date_of_birth(df_copy)

# Check the processed 'Date of Birth' column
df_copy['date_of_birth'].head()
```

```python
# Function to extract the date part (YYYY-MM-DD) from a datetime string
def extract_date_part(date_str):
    try:
        # Check if the string has a time component and extract only the date part
        if isinstance(date_str, str) and ' ' in date_str:
            # Extract the part before the space (date part)
            date_str = date_str.split(' ')[0]
        # Return the date string
        return date_str
    except Exception as e:
        return date_str

# Normalize the date format: MM/DD/YYYY -> YYYY-MM-DD
def normalize_date_format(date_str):
    try:
        # If format is like 'MM/DD/YYYY', change it to 'YYYY-MM-DD'
        if re.match(r'\d{2}/\d{2}/\d{4}', date_str):  # Matches MM/DD/YYYY
            return pd.to_datetime(date_str, format='%m/%d/%Y').strftime('%Y-%m-%d')
        # If already in 'YYYY-MM-DD', return as is (drop time part)
        return pd.to_datetime(date_str).strftime('%Y-%m-%d')
    except Exception as e:
        # Return the original string if any error occurs
        return date_str

# List of columns to process
date_columns = ['learner_signup_datetime', 'opportunity_end_date', 'entry_created_at', 'apply_date', 'opportunity_start_date']

# Loop through each column and apply the functions
for col in date_columns:
    # Ensure the column is in string format
    df_copy[col] = df_copy[col].astype(str)

    # Step 1: Extract the date part (without time)
    df_copy[col] = df_copy[col].apply(extract_date_part)

    # Step 2: Normalize the date format: MM/DD/YYYY -> YYYY-MM-DD
    df_copy[col] = df_copy[col].apply(normalize_date_format)

    # Step 3: Convert the column to datetime, catching invalid entries
    df_copy[col] = pd.to_datetime(df_copy[col], errors='coerce')

# Check the processed columns
df_copy[date_columns].head()
```

# **Feature Engineering**

# creating new features

#### Age Calculation: Age from Date of Birth

```python
df_copy['age'] = (datetime.now() - df_copy['date_of_birth']).dt.days // 365
df_copy['age'].head()
```

#### Opportunity Duration: Duration between Opportunity Start and End Date

```python
df_copy['opportunity_duration'] = (df_copy['opportunity_end_date'] - df_copy['opportunity_start_date']).dt.days
df_copy['opportunity_duration'].head()
```

# Extracting Useful Components

#### Extract Month, Year, and Day for Each datetime Column

```python
# List of datetime columns
datetime_columns = ['learner_signup_datetime', 'opportunity_end_date', 'date_of_birth', 'entry_created_at', 'apply_date', 'opportunity_start_date']

# Loop through each datetime column
for col in datetime_columns:
    # Extract Year, Month, Day and create new columns
    df_copy[f'{col}_year'] = df_copy[col].dt.year
    df_copy[f'{col}_month'] = df_copy[col].dt.month
    df_copy[f'{col}_day'] = df_copy[col].dt.day

# Check the updated dataframe with the new columns
df_copy[datetime_columns + [f'{col}_year' for col in datetime_columns] + [f'{col}_month' for col in datetime_columns] + [f'{col}_day' for col in datetime_columns]].head()
df_copy.head()
```

#### Opportunity Engagement: Days between Apply Date and Opportunity Start Date

```python
df_copy['engagement_time'] = (df_copy['apply_date'] - df_copy['opportunity_start_date']).dt.days
df_copy['engagement_time'].head(20)
```

# Combining Features

#### Interaction Feature: Multiply Opportunity Duration and Age to get an interaction score


```python
# Now you can proceed with the calculation of 'duration_age_interaction'
df_copy['duration_age_interaction'] = df_copy['opportunity_duration'] * df_copy['age']
df_copy['duration_age_interaction']
```

#### Engagement Scores: Weighted average of Opportunity Duration, Age, and Engagement Time


```python
df_copy['engagement_score'] = (0.4 * df_copy['opportunity_duration']) + (0.3 * df_copy['age']) + (0.3 * df_copy['engagement_time'])
df_copy['engagement_score']

#The engagement score is calculated using a weighted sum of three different columns:

# opportunity_duration (40% weight - 0.4)
# age (30% weight - 0.3)
# engagement_time (30% weight - 0.3)

# How It Works:
# Each column value is multiplied by its assigned weight.
# The weighted values are summed up to get the final engagement_score.
# The new score is stored in a new column engagement_score in df_copy.
```

### drop_off

```python
# Define the drop-off statuses
drop_off_statuses = ['Dropped Out', 'Withdraw']

# Define a function to classify the outcome
def classify_drop_off(status_description):
    if status_description in drop_off_statuses:
        return 1  # Drop-off
    else:
        return 0  # No drop-off

# Apply this function to create a new binary column 'drop_off'
df_copy['drop_off'] = df_copy['status_description'].apply(classify_drop_off)
```
