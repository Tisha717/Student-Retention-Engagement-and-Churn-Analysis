### Opportunity Duration vs. Completion Rate

```python
# unique values in "status_description" column
df.status_description.unique()
```

##### Creating "completion_rate" Column Using "status_description"

- **1 = Successfully Completed**  
  - `status_description == "Rewards-Award"` OR `"Started"` OR `"Applied"` OR `"Team Allocated"`

- **0 = Not Completed**  
  - `status_description == "Dropped out"` OR `"Withdraw"` OR `"Rejected"` OR `"Waitlisted"`


```python
# Completion status (1 = Completed, 0 = Not Completed)
df["completion_status"] = df["status_description"].apply(lambda x: 1 if x in ["Rewards-Award", "Started", "Team Allocated"] else 0)
```

```python
# Calculating completion rate per opportunity duration
completion_summary = df.groupby("opportunity_duration")["completion_status"].mean().reset_index()
completion_summary.columns = ["Opportunity Duration (Days)", "Average Completion Rate"]
```

```python
# result
print(completion_summary.head(10))  # first 10 rows
```

```python
# visualization
# Method-1
# Line Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x="Opportunity Duration (Days)", y="Average Completion Rate", data=completion_summary, marker="o", color="b")

plt.title("Opportunity Duration vs. Completion Rate", fontsize=14)
plt.xlabel("Opportunity Duration (Days)", fontsize=12)
plt.ylabel("Average Completion Rate", fontsize=12)
plt.axhline(y=1.0, color='gray', linestyle='--', alpha=0.5)  # Reference line for 100% completion
plt.axhline(y=0.5, color='red', linestyle='--', alpha=0.5)   # Reference line for 50% completion
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Method-2
# Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x="Opportunity Duration (Days)", y="Average Completion Rate", data=completion_summary, color="blue", alpha=0.6)

plt.title("Scatter Plot: Opportunity Duration vs. Completion Rate", fontsize=14)
plt.xlabel("Opportunity Duration (Days)", fontsize=12)
plt.ylabel("Average Completion Rate", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

#### **Insights for "Opportunity Duration vs. Completion Rate"**

#####  **1 Shorter-duration opportunities have higher completion rates**
- Opportunities with **0 to 7 days** have the **highest completion rates (above 94%)**, with **some reaching 100%**.
- This suggests that **shorter commitments are more manageable** and **learners are more likely to complete them**.
- **Example:**  
  - **0 days  97.1% completion**  
  - **7 days  100% completion**  

#####  **2 Mid-duration opportunities show mixed trends**
- **Opportunities between 21 to 60 days** exhibit **fluctuations** in completion rates.  
- Some have **high completion rates (e.g., 45 days - 100%)**, while others drop significantly (e.g., **49 days - 37.8%**, **60 days - 0%**).
- This indicates that **engagement strategies may play a role in mid-term opportunities**.  

#####  **3 Very long-duration opportunities (>100 days) face severe drop-offs**
- Many **longer-duration opportunities (100+ days) show 0% completion**.
- However, **some exceptions exist**, where engagement strategies might have been successful (e.g., **468 days - 96.2%, 604 days - 98.9%**).
- This suggests that **longer programs need structured support to maintain engagement**.  

#####  **4 Negative Opportunity Duration - Possible Data Issue**
- **Some opportunities have negative durations (e.g., -313, -69 days) with 0% completion**.
- This could indicate **data entry errors** or **misalignment between start and apply dates**.
- **Action:** Investigate these cases to ensure data accuracy.  

---

#### **Actionable Recommendations **
 **Optimize shorter opportunities**: Promote and expand opportunities with **<7 days duration**, as they have near-perfect completion rates.  
 **Enhance engagement for mid-term programs**: Identify what makes **some 21-60 day opportunities successful while others struggle** and apply best practices.  
 **Improve support for long-term opportunities**: Implement structured mentoring, milestone tracking, and incentives to **increase engagement in programs >100 days**.  
 **Fix data inconsistencies**: Review and clean negative-duration records to ensure meaningful analysis.

### Drop-off Rate by Opportunity Category

##### Creating `"drop_off_status"` Column Similar to `"completion_status"` Column Using `"status_description"`
- **1 = Dropped Off** (if `status_description` is **"Waitlisted"**, **"Withdraw"**, **"Dropped Out"**, or **"Rejected"**)
- **0 = Not Dropped Off** (if `status_description` is **"Started"**, **"Team Allocated"**, **"Rewards Award"**, or **"Applied"**)


```python
# Defining drop-off statuses
drop_off_statuses = ["Dropped out", "Rejected", "Withdraw","Waitlisted"]

# Creating drop_off_status column (1 = dropped off, 0 = did not drop off)
df["drop_off_status"] = np.where(df["status_description"].isin(drop_off_statuses), 1, 0)

# Calculating drop-off rate per opportunity category
drop_off_summary = df.groupby("opportunity_category")["drop_off_status"].mean().reset_index()

# Renaming columns for better clarity
drop_off_summary.columns = ["Opportunity Category", "Drop-off Rate"]
```

```python
# result
print(drop_off_summary)
```

```python
# Visualization
plt.figure(figsize=(10, 5))
sns.barplot(x="Opportunity Category", y="Drop-off Rate", data=drop_off_summary, palette="coolwarm_r")
plt.title("Drop-off Rate Across Opportunity Categories")
plt.xlabel("Opportunity Category")
plt.ylabel("Drop-off Rate")
plt.xticks(rotation=45)
plt.show()
```

####  Drop-off Rate by Opportunity Type - Insights

#####  Key Insights:
1. **Internships have the highest drop-off rate (66.9%)**   
   - This suggests that many learners **struggle to complete internships**, possibly due to **long durations, workload, or mismatch in expectations**.  
   - A deeper analysis into **drop-off reasons** can help in **improving retention**.  

2. **Courses have a moderate drop-off rate (6.4%)**   
   - This indicates some learners **start but do not finish courses**.  
   - Possible reasons could be **lack of engagement, difficulty level, or competing priorities**.  
   - Enhancing course **content, flexibility, or support** may help reduce drop-offs.  

3. **Competitions (0.2%) and Events (0.4%) have very low drop-off rates**   
   - These opportunities likely **require less commitment**, making them **easier to complete**.  

4. **Engagement-based opportunities have a 0% drop-off rate**   
   - Indicates **high participation & retention**suggesting that these opportunities **align well with learners' interests and availability**.  

#####  **Actionable Recommendations**
- **Improve internship retention** by:
  - Introducing **check-in sessions**   
  - Offering **support resources**   
  - Matching learners with **better-suited opportunities**   

- **Enhance course completion rates** through:
  - **Gamification & rewards**   
  - **Personalized learning paths**   
  - **Better progress tracking & reminders**   

- **Encourage more signups for successful opportunity types** like **Engagements & Competitions**

### Application & Signup Trends Over Time

```python
# Ensuring both columns are in datetime format
df["learner_signup_datetime"] = pd.to_datetime(df["learner_signup_datetime"])
df["apply_date"] = pd.to_datetime(df["apply_date"])

# Extracting Year-Month
df["Signup Year-Month"] = df["learner_signup_datetime"].dt.to_period("M")
df["Apply Year-Month"] = df["apply_date"].dt.to_period("M")

# Aggregate counts for both
signup_trends = df.groupby("Signup Year-Month").size().reset_index(name="Signup Count")
apply_trends = df.groupby("Apply Year-Month").size().reset_index(name="Application Count")

# Converting Period to String for plotting
signup_trends["Signup Year-Month"] = signup_trends["Signup Year-Month"].astype(str)
apply_trends["Apply Year-Month"] = apply_trends["Apply Year-Month"].astype(str)

# Merging both trends on Year-Month for alignment
trend_summary = pd.merge(signup_trends, apply_trends, left_on="Signup Year-Month", right_on="Apply Year-Month", how="outer").fillna(0)
```
