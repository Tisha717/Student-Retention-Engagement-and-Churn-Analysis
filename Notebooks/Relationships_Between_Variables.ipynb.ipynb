## Relationships between variables


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
from IPython.display import display
from datetime import datetime
```

```python
df = pd.read_csv('/content/Cleaned_Dataset_Week1.csv')
```

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='engagement_score', data=df)
plt.title('Relationship between Age and Engagement Score')
plt.xlabel('Age')
plt.ylabel('Engagement Score')
plt.show()

# Calculate the correlation coefficient
correlation = df['age'].corr(df['engagement_score'])
print(f"Correlation between Age and Engagement Score: {correlation}")
```

```python
# Box plot to compare engagement scores across genders
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='engagement_score', data=df)
plt.title('Relationship between Gender and Engagement Score')
plt.xlabel('Gender')
plt.ylabel('Engagement Score')
plt.show()

# Calculate the average engagement score for each gender
avg_engagement_by_gender = df.groupby('gender')['engagement_score'].mean()
print(f"Average Engagement Score by Gender: \n{avg_engagement_by_gender}")
```

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='opportunity_duration', y='engagement_score', data=df)
plt.title('Relationship between Opportunity Duration and Engagement Score')
plt.xlabel('Opportunity Duration')
plt.ylabel('Engagement Score')
plt.show()

# Calculate the correlation coefficient
correlation = df['opportunity_duration'].corr(df['engagement_score'])
print(f"Correlation between Opportunity Duration and Engagement Score: {correlation}")
```

```python
# Convert 'apply_date' to datetime objects if it's not already
df['apply_date'] = pd.to_datetime(df['apply_date'])

# Create a new column for the day of the week the application was submitted
df['apply_dayofweek'] = df['apply_date'].dt.dayofweek

# Box plot to compare engagement scores across different days of the week
plt.figure(figsize=(8, 6))
sns.boxplot(x='apply_dayofweek', y='engagement_score', data=df)
plt.title('Relationship between Apply Date and Engagement Score')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Engagement Score')
plt.show()
```

```python
import statsmodels.formula.api as smf

# Create an interaction term
df['age_duration_interaction'] = df['age'] * df['opportunity_duration']

# Fit a regression model with the interaction term
model = smf.ols('engagement_score ~ age + opportunity_duration + age_duration_interaction', data=df).fit()

# Print the model summary
print(model.summary())
```

```python


# Relationship with Opportunity Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='opportunity_category', y='engagement_score', data=df)
plt.title('Relationship between Opportunity Category and Engagement Score')
plt.xlabel('Opportunity Category')
plt.ylabel('Engagement Score')
plt.xticks(rotation=45, ha='right')
plt.show()

# Relationship with Status Code
plt.figure(figsize=(12, 6))
sns.boxplot(x='status_code', y='engagement_score', data=df)
plt.title('Relationship between Status Code and Engagement Score')
plt.xlabel('Status Code')
plt.ylabel('Engagement Score')
plt.xticks(rotation=45, ha='right')
plt.show()

# Relationship with Country
plt.figure(figsize=(12, 6))
sns.boxplot(x='country', y='engagement_score', data=df)
plt.title('Relationship between Country and Engagement Score')
plt.xlabel('Country')
plt.ylabel('Engagement Score')
plt.xticks(rotation=45, ha='right')
plt.show()

# Relationship with Learner Signup Datetime
df['learner_signup_datetime'] = pd.to_datetime(df['learner_signup_datetime'])
df['signup_dayofweek'] = df['learner_signup_datetime'].dt.dayofweek

plt.figure(figsize=(12, 6))
sns.boxplot(x='signup_dayofweek', y='engagement_score', data=df)
plt.title('Relationship between Learner Signup Datetime and Engagement Score')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Engagement Score')
plt.show()
```
