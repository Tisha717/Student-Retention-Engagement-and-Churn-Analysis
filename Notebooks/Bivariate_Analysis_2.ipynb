```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Set plot style
sns.set(style="whitegrid")

# Define variable pairs for scatter plots
scatter_pairs = [
    ("age", "engagement_score", "Age vs. Engagement Score"),
    ("opportunity_duration", "engagement_time", "Opportunity Duration vs. Engagement Time"),
    ("status_code", "engagement_score", "Status Code vs. Engagement Score")
]
```

```python
# Create scatter plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (x, y, title) in zip(axes, scatter_pairs):
    sns.scatterplot(data=df, x=x, y=y, alpha=0.5, ax=ax)
    ax.set_title(title)

```

```python
plt.tight_layout()
plt.show()
# Set plot style
sns.set(style="whitegrid")

# Define variable pairs for scatter plots
scatter_pairs = [
    ("age", "engagement_score", "Age vs. Engagement Score"),
    ("opportunity_duration", "engagement_time", "Opportunity Duration vs. Engagement Time"),
    ("status_code", "engagement_score", "Status Code vs. Engagement Score")
]

```
